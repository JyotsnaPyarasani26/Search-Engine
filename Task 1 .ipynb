### WEB CRAWLER 

```python
import time
import tqdm
Authors = []
Publication_year = []
Title = []
Publication_page_link = []
Authors_profile_page_link = []

# Loop through multiple pages of the website
for i in tqdm.tnrange(0, 7):
    URL = 'https://pureportal.coventry.ac.uk/en/organisations/centre-global-learning/publications/?page={}'.format(i)
    page = requests.get(URL)
    htmlCode = page.text

    soup = BeautifulSoup(htmlCode)

    # Extract data for each publication on the current page
    for x in soup.find_all('div', attrs={'class': 'result-container'}):
        # Extract Authors
        authors = x.find_all('a', attrs={'class': 'person'})
        if authors is None:
            Authors.append(np.NaN)
        else:
            Authors.append([author.text for author in authors])

        # Extract Publication Year
        publication_year = x.find('span', attrs={'class': 'date'})
        if publication_year is None:
            Publication_year.append(np.NaN)
        else:
            Publication_year.append(publication_year.text)

        # Extract Title
        title = x.find('h3', attrs={'class': 'title'})
        if title is None:
            Title.append(np.NaN)
        else:
            Title.append(title.text)

        # Extract Publication Page Link
        publication_page_link = x.find('a', attrs={'class': 'link'})
        if publication_page_link is None:
            Publication_page_link.append(np.NaN)
        else:
            Publication_page_link.append(publication_page_link['href'])

        # Extract Authors Profile Page Links
        authors_profile_page_link = x.find_all('a', attrs={'class': 'link person'})
        if authors_profile_page_link is None:
            Authors_profile_page_link.append(np.NaN)
        else:
            authors_profile_page_links_list = []
            for author_profile_link in authors_profile_page_link:
                authors_profile_page_links_list.append(author_profile_link['href'])
            Authors_profile_page_link.append(authors_profile_page_links_list)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download the required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Step 1: Convert to lowercase
df1['Authors'] = df1['Authors'].str.lower()
df1['Title'] = df1['Title'].str.lower()

# Step 2: Tokenize the strings into individual words
df1['Authors'] = df1['Authors'].apply(word_tokenize)
df1['Title'] = df1['Title'].apply(word_tokenize)

# Step 3: Remove punctuation and stopwords
stop_words = set(stopwords.words('english'))

def remove_punctuation_and_stopwords(tokens):
    return [token for token in tokens if token.isalpha() and token not in stop_words]

df1['Authors'] = df1['Authors'].apply(remove_punctuation_and_stopwords)
df1['Title'] = df1['Title'].apply(remove_punctuation_and_stopwords)

# Step 4: Apply stemming 
stemmer = PorterStemmer()

def apply_stemming(tokens):
    return [stemmer.stem(token) for token in tokens]

df1['Authors'] = df1['Authors'].apply(apply_stemming)
df1['Title'] = df1['Title'].apply(apply_stemming)

```

### INDEXING

```python
import pandas as pd
from collections import defaultdict

# the DataFrame df1 with 'Authors' and 'Title' columns preprocessed

# Step 1: Initialize an empty index
index = defaultdict(list)

# Step 2: Iterate through the DataFrame and populate the index
for idx, row in df2.iterrows():
    doc_id = idx  # The index of the row in the DataFrame serves as the document ID
    authors = row['Authors']
    title = row['Title']
    
    # Check if authors and title are not NaN (not missing values)
    if pd.notna(authors):
        authors = authors.split()  # Split the authors into individual words
    else:
        authors = []
        
    if pd.notna(title):
        title = title.split()  # Split the title into individual words
    else:
        title = []
    
    # Update the index with each word and its corresponding document ID
    for word in authors + title:
        index[word].append(doc_id)

# Now, the 'index' dictionary contains words as keys and a list of corresponding document IDs as values.


```

### User Query Processing

```python
import pandas as pd
from collections import defaultdict

# I have the DataFrame df1 with 'Authors' and 'Title' columns preprocessed
# I have already created the index as shown in the previous code

# Function to perform a search and return the matching documents
def search(query):
    query_terms = query.lower().split()  # Preprocess the query in the same way as the documents
    matching_doc_ids = set()  # Use a set to store the matching document IDs to avoid duplicates
    
    # Iterate through each query term and find matching document IDs from the index
    for term in query_terms:
        doc_ids = index.get(term, [])  # Get the list of document IDs for the query term
        matching_doc_ids.update(doc_ids)  # Add the document IDs to the set of matching IDs
    
    # Retrieve the matching documents from the DataFrame
    matching_documents = df2.loc[list(matching_doc_ids)]
    
    return matching_documents

```

```python
# Example usage:
search_query = "english"  # User's search query
matching_results = search(search_query)

# The 'matching_results' DataFrame will contain all the documents that contain the words "machine" and "learning" in either 'Authors' or 'Title' columns.

# You can display the relevant fields (Authors, Publication_year, Title, Publication_page_link, Authors_profile_page_link) for each match as required.
```
